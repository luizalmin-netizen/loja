Now, we can load these variables using `python-dotenv` and access them using `os.getenv()`.

```python
from dotenv import load_dotenv
import os

load_dotenv()

trello_api_key = os.getenv('TRELLO_API_KEY')
trello_token = os.getenv('TRELLO_TOKEN')

print(f"Trello API Key: {trello_api_key}")
print(f"Trello Token: {trello_token}")
```

```python
import requests
import json

# Trello API endpoint for current member's boards
boards_url = f"https://api.trello.com/1/members/me/boards?key={trello_api_key}&token={trello_token}"

try:
    response = requests.get(boards_url)
    response.raise_for_status() # Raise an exception for HTTP errors (4xx or 5xx)
    boards = response.json()

    print("Your Trello Boards:")
    if boards:
        for board in boards:
            print(f"- {board['name']} (ID: {board['id']})")
    else:
        print("No boards found. Make sure your API key and token are correct and have the necessary permissions.")

except requests.exceptions.RequestException as e:
    print(f"Error making request to Trello API: {e}")
except json.JSONDecodeError:
    print("Error decoding JSON response from Trello API.")
except Exception as e:
    print(f"An unexpected error occurred: {e}")
```
